```python
#QUESTION-1-Top 7 RICE PRODUCTION State Data(Bar_plot)
rice_production_by_state = df.groupby('state_name')['rice_production_(1000_tons)'].sum().reset_index()
top_7_states = rice_production_by_state.nlargest(7,'rice_production_(1000_tons)')
print(top_7_states)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
colors = sns.color_palette("magma", len(top_7_states))
plt.figure(figsize=(10, 5),facecolor='lavender')
ax = plt.axes()
ax.set_facecolor("lightgrey")
plt.bar(top_7_states['state_name'], top_7_states['rice_production_(1000_tons)'], color=colors)
plt.xlabel("State", fontsize=12)
plt.ylabel("Total Rice Production (1000 tons)", fontsize=12)
plt.title("Top 7 Rice-Producing States in India", fontsize=14)
handles = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(top_7_states))]
labels = top_7_states['state_name'].tolist()
plt.legend(handles, labels,title="States",loc="upper right",fontsize=10, facecolor='white', edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

```python
#QUESTION-2-Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
wheat_production_by_state=df.groupby('state_name')['wheat_production_(1000_tons)'].sum().reset_index()
top_5_states=wheat_production_by_state.nlargest(5,'wheat_production_(1000_tons)')
total_wheat_production = wheat_production_by_state['wheat_production_(1000_tons)'].sum()
top_5_states['percentage_contribution'] = (top_5_states['wheat_production_(1000_tons)'] / total_wheat_production) * 100
other_states_production = total_wheat_production - top_5_states['wheat_production_(1000_tons)'].sum()
other_states_percentage = (other_states_production / total_wheat_production) * 100
other_states_row = pd.DataFrame({'state_name': ['Other States'], 'wheat_production_(1000_tons)': [other_states_production], 'percentage_contribution': [other_states_percentage]})
final_data = pd.concat([top_5_states, other_states_row])
print(top_5_states)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5),facecolor='pink')
ax = plt.axes()
ax.set_facecolor("lavenderblush")
colors = sns.color_palette("viridis", len(top_5_states))
plt.bar(top_5_states['state_name'],top_5_states['wheat_production_(1000_tons)'],color=colors)
plt.xlabel("State",fontsize=12)
plt.ylabel("Total Wheat Production (1000 tons)",fontsize=12)
plt.title("Top 5 Wheat-Producing States in India",fontsize=14)
handles = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(top_5_states))]
labels = top_5_states['state_name'].tolist()
plt.legend(handles, labels,title="States",loc="upper right",fontsize=10, facecolor='white', edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

```python
#QUESTION-14-Impact of Area Cultivated on Production (Rice, Wheat, Maize)
impact_data = df.groupby(["state_name", "year"]).agg({"rice_area_(1000_ha)": "sum","rice_production_(1000_tons)": "sum",
"wheat_area_(1000_ha)": "sum","wheat_production_(1000_tons)": "sum",
"maize_area_(1000_ha)": "sum","maize_production_(1000_tons)": "sum"}).reset_index()
print(impact_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
melted_data = impact_data.melt(id_vars=["state_name", "year"], value_vars=["rice_area_(1000_ha)", "wheat_area_(1000_ha)", "maize_area_(1000_ha)",
"rice_production_(1000_tons)", "wheat_production_(1000_tons)", "maize_production_(1000_tons)"], var_name="Type", value_name="Value")
melted_data["Crop"] = melted_data["Type"].str.extract(r"(rice|wheat|maize)", expand=False).str.capitalize()
melted_data["Metric"] = melted_data["Type"].apply(lambda x: "Area" if "area" in x else "Production")
pivot_data = melted_data.pivot(index=["state_name", "year", "Crop"], columns="Metric", values="Value").reset_index()
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pivot_data,x="Area",y="Production",hue="Crop",style="state_name",size="Production",sizes=(50, 300),palette="Set1")
plt.xlabel("Cultivated Area (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.title("Impact of Cultivated Area on Production")
plt.legend(title="Crop Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
#QUESTION-1-Year-wise Trend of Rice Production Across States (Top 3)
mycursor.execute("""select year, state_name, sum(`rice_production_(1000_tons)`) as total_rice_production
from AGRI_DATA.CROP_DETAILS
where state_name in (select state_name from AGRI_DATA.CROP_DETAILS group by state_name order by sum(`rice_production_(1000_tons)`) desc limit 3)
group by year, state_name
order by year desc, total_rice_production desc""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-2-Top 5 Districts by Wheat Yield Increase Over the Last 5 Years
mycursor.execute("""select dist_name, (max(`wheat_yield_(kg_per_ha)`)-min(`wheat_yield_(kg_per_ha)`)) as wheat_yield_increase
from AGRI_DATA.CROP_DETAILS
where year>=(select max(year) from AGRI_DATA.CROP_DETAILS)-4
group by dist_name
order by wheat_yield_increase desc limit 5""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-3-States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
mycursor.execute("""with oilseed_growth as (select
state_name,
min(case when year = 2013 then `oilseeds_production_(1000_tons)` end) as prod_2013,
max(CASE WHEN year = 2017 THEN `oilseeds_production_(1000_tons)` end) as prod_2017
from AGRI_DATA.CROP_DETAILS
group by state_name)
select state_name, prod_2013, prod_2017,
((prod_2017 - prod_2013) * 100.0 / nullif(prod_2013, 0)) as growth_rate
from oilseed_growth
order by growth_rate desc limit 3""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
